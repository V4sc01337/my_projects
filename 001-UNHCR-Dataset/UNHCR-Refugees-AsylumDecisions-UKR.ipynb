## Plotting

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# Create an axes object 
axes = plt.gca()

# pass the axes object to plot function
#df.plot(kind='line', x='Year', y='Recognized decisions', ax=axes);
#df.plot(kind='line', x='Year', y='Complementary protection', ax=axes);
#df.plot(kind='line', x='Year', y='Rejected decisions', ax=axes);
#df.plot(kind='line', x='Year', y='Otherwise closed', ax=axes);
df.plot(kind='line', x='Year', y='Total decisions', ax=axes);
```

```python
# Group Data by Year 
grouped = df.groupby('Year').sum(numeric_only=True)
```

```python
# Line Diagram 
grouped.plot(kind='line', marker='o', linestyle='dashed', figsize=(15,5))
plt.xlabel('Year')
plt.ylabel('Total')
plt.title('Asylum Decisions over Time (Ukraine UNHCR Data)')
plt.show()
```

```python
# Export new *.csv
grouped.to_csv('UNHCR-Asylum-decisions.csv')
```
